```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Forecasting the spatial distribution of biomass for 2018 and 2019 using simple linear regression
years = [str(year) for year in range(2010, 2018)]
biomass_history["2018"] = 0
biomass_history["2019"] = 0

for idx, row in biomass_history.iterrows():
    X = [[year] for year in range(2010, 2018)]
    y = row[years].values
    model = LinearRegression().fit(X, y)
    biomass_history.at[idx, "2018"] = model.predict([[2018]])[0]
    if biomass_history.at[idx, "2018"] < 0:
      biomass_history.at[idx, "2018"] = 0
    biomass_history.at[idx, "2019"] = model.predict([[2019]])[0]
    if biomass_history.at[idx, "2019"] < 0:
      biomass_history.at[idx, "2019"] = 0

# Displaying the forecasted biomass for 2018 and 2019
biomass_history[["Index", "Latitude", "Longitude", "2018", "2019"]].head()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Function to forecast biomass using ARIMA model for a given site
def forecast_biomass(site_data):
    model = ARIMA(site_data, order=(5,1,0),enforce_stationarity=False)
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=2)
    return forecast

# Forecast biomass for each site using ARIMA
biomass_history1["2018_ARIMA"] = 0
biomass_history1["2019_ARIMA"] = 0

for idx, row in biomass_history1.iterrows():
    try:
        forecast = forecast_biomass(row[years].values)
        biomass_history1.at[idx, "2018_ARIMA"] = forecast[0]
        biomass_history1.at[idx, "2019_ARIMA"] = forecast[1]
    except:
        # If ARIMA fails for some rows, retain the linear regression forecast
        biomass_history1.at[idx, "2018_ARIMA"] = biomass_history.at[idx, "2018"]
        biomass_history1.at[idx, "2019_ARIMA"] = biomass_history.at[idx, "2019"]

# Display the forecasted biomass using ARIMA for 2018 and 2019
```

```python
for idx, row in biomass_history1.iterrows():
    # Check if the ARIMA prediction is greater than 0
    if row["2018_ARIMA"] < 0:
        # Check the linear regression prediction
        linear_regression_value = biomass_history.at[idx, "2018"]
        if linear_regression_value >= 0:
            # Use the linear regression value
            biomass_history1.at[idx, "2018_ARIMA"] = linear_regression_value
        else:
            # Set production as 0
            biomass_history1.at[idx, "2018_ARIMA"] = 0
    if row["2019_ARIMA"] < 0:
        # Check the linear regression prediction
        linear_regression_value = biomass_history.at[idx, "2018"]
        if linear_regression_value >= 0:
            # Use the linear regression value
            biomass_history1.at[idx, "2018_ARIMA"] = linear_regression_value
        else:
            # Set production as 0
            biomass_history1.at[idx, "2018_ARIMA"] = 0
```

```python
biomass_history1["Total_2018_2019_ARIMA"] = biomass_history1["2018_ARIMA"] + biomass_history1["2019_ARIMA"]
```

```python
# Parameters
max_depot_capacity = 20000.00
max_depots = 25

# List to store depot locations and the harvesting sites they serve
depots = []

for _ in range(max_depots):
    # Start with the top site with the highest remaining biomass
    current_site = remaining_biomass.iloc[0]

    # Get the current site's index as an integer
    current_site_index = int(current_site["Index"])

    if current_site_index in distance_matrix.index:
        # Determine nearby sites based on distance
        nearby_sites = distance_matrix.iloc[current_site_index].nsmallest(101).index[1:]

    # Create a depot at the current site and start serving nearby sites
        depot = {"location": (current_site["Latitude"], current_site["Longitude"]),
                 "served_sites": [current_site_index],
                 "total_biomass": current_site["Total_2018_2019_ARIMA"]}

        for site in nearby_sites:
          site = int(site)
          if site in remaining_biomass["Index"].values:
            site_biomass = remaining_biomass[remaining_biomass["Index"] == site]["Total_2018_2019_ARIMA"].values[0]
            if (depot["total_biomass"] + site_biomass) <= max_depot_capacity:
              depot["served_sites"].append(site)
              depot["total_biomass"] += site_biomass

        # Add the depot to the list
        depots.append(depot)

        # Remove served sites from the remaining biomass dataframe
        remaining_biomass = remaining_biomass[~remaining_biomass["Index"].isin(depot["served_sites"])]
        remaining_biomass
        # If all sites are served, break the loop
        if remaining_biomass.empty:
            break
    else:
        # Handle the case where the current site index does not exist in the distance matrix
        print(f"Index {current_site_index} does not exist in the distance matrix.")
        break

for depot in depots:
  print(f"Depot at Location {depot['location']}")
  print(f"Served Sites: {depot['served_sites']}")
  print(f"Total Biomass Served: {depot['total_biomass']}")
  print("\n")
```

```python
import pandas as pd

# Load the necessary datasets
depots_df = pd.read_csv("depots.csv")
biomass_predictions = top_sites1[["Index", "Latitude", "Longitude", "2018_ARIMA", "2019_ARIMA", "Total_2018_2019_ARIMA"]].head(10)

remaining_biomass_df = pd.read_csv("remaining_biomass.csv")

# Identify the depot with the highest amount of biomass
top_depot = depots_df.sort_values(by="total_biomass", ascending=False).iloc[0]

top_depot
```

```python
# Constants
max_biorefinery_capacity = 100000.00
max_biorefineries = 5

# List to store biorefinery locations and the depots they serve
biorefineries = []

# Start with the depot with the highest biomass and assign it to a biorefinery
biorefinery = {
    "location": top_depot["location"],
    "served_depots": [top_depot["location"]],
    "total_biomass": top_depot["total_biomass"]
}

# Add the biorefinery to the list
biorefineries.append(biorefinery)

# Remove the assigned depot from the depots dataframe
depots_df = depots_df[depots_df["location"] != top_depot["location"]]

biorefinery
```

```python
import numpy as np

# Function to calculate Euclidean distance between two points
def euclidean_distance(point1, point2):
    return np.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)

# Include the current biorefinery location (which was also the top depot) in the distance matrix calculation
depots_df = pd.concat([depots_df, top_depot.to_frame().T])

# Calculate the distance matrix for depots
depot_locations = depots_df[["latitude", "longitude"]].values
num_depots = len(depot_locations)
distance_matrix = np.zeros((num_depots, num_depots))

for i in range(num_depots):
    for j in range(num_depots):
        distance_matrix[i, j] = euclidean_distance(depot_locations[i], depot_locations[j])

# Convert the numpy matrix to a pandas DataFrame for easier indexing
distance_df = pd.DataFrame(distance_matrix, index=depots_df["location"].values, columns=depots_df["location"].values)

# Get the distances of nearby depots to our current biorefinery
current_location = biorefinery["location"]
nearby_depots = distance_df[current_location].nsmallest(31)

nearby_depots.head()  # Displaying the top 5 closest depots for clarity
```

```python
# Add nearby depots to the current biorefinery until its capacity is reached
for depot_location, distance in nearby_depots.items():
    if depot_location != current_location:  # Exclude the biorefinery itself
        depot_biomass = depots_df[depots_df["location"] == depot_location]["total_biomass"].values[0]
        if (biorefinery["total_biomass"] + depot_biomass) <= max_biorefinery_capacity:
            biorefinery["served_depots"].append(depot_location)
            biorefinery["total_biomass"] += depot_biomass
            # Remove the assigned depot from the depots dataframe
            depots_df = depots_df[depots_df["location"] != depot_location]
        else:
            break  # Stop if biorefinery capacity is reached

biorefinery
```

```python
# Iterate to create additional biorefineries and assign them depots
while len(biorefineries) < max_biorefineries and not depots_df.empty:
    # Start with the depot with the highest remaining biomass
    top_depot = depots_df.sort_values(by="total_biomass", ascending=False).iloc[0]
    biorefinery = {
        "location": top_depot["location"],
        "served_depots": [top_depot["location"]],
        "total_biomass": top_depot["total_biomass"]
    }
    # Remove the assigned depot from the depots dataframe
    depots_df = depots_df[depots_df["location"] != top_depot["location"]]

    # Check nearby depots and assign them to the biorefinery
    nearby_depots = distance_df[biorefinery["location"]].nsmallest(num_depots)
    for depot_location, distance in nearby_depots.items():
        if depot_location != biorefinery["location"]:  # Exclude the biorefinery itself
            if depot_location in depots_df["location"].values:
                depot_biomass = depots_df[depots_df["location"] == depot_location]["total_biomass"].values[0]
                if (biorefinery["total_biomass"] + depot_biomass) <= max_biorefinery_capacity:
                    biorefinery["served_depots"].append(depot_location)
                    biorefinery["total_biomass"] += depot_biomass
                    # Remove the assigned depot from the depots dataframe
                    depots_df = depots_df[depots_df["location"] != depot_location]
                else:
                    break  # Stop if biorefinery capacity is reached

    # Add the new biorefinery to the list
    biorefineries.append(biorefinery)

# Display the created biorefineries
biorefineries
```

```python
# Convert biorefineries list to DataFrame
biorefineries_df = pd.DataFrame(biorefineries)
biorefineries_df
```

```python
# Convert the string representation in "location" column back to tuple objects
biorefineries_df['location'] = biorefineries_df['location'].apply(eval)

# Split the "location" column into "latitude" and "longitude" columns for biorefineries
biorefineries_df[['latitude', 'longitude']] = pd.DataFrame(biorefineries_df['location'].tolist(), columns=['latitude', 'longitude'])

# Display the updated biorefineries dataframe
biorefineries_df
```
